```python
from newspaper import Article
article = Article('https://en.wikipedia.org/wiki/Economy_of_India')
article.download()
article.parse()
```

```python
article.text
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(background_color="white", max_words=2000,
               stopwords=STOPWORDS, max_font_size=256,
               random_state=42, width=1000, height=500)
wc.generate(article.text)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import numpy as np
```

```python
mask = np.array(Image.open('images.jpeg'))
wc = WordCloud(stopwords=STOPWORDS,
               mask=mask, background_color="white",
               max_words=2000, max_font_size=256,
               random_state=42, width=mask.shape[1],
               height=mask.shape[0])
wc.generate(article.text)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

In case of presentations, custom word clouds shall come in handy to convey information better.
